```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
breast_cancer_wisconsin_diagnostic = fetch_ucirepo(id=17) 
  
# data (as pandas dataframes) 
Id = breast_cancer_wisconsin_diagnostic.data.ids
Features = breast_cancer_wisconsin_diagnostic.data.features 
Target = breast_cancer_wisconsin_diagnostic.data.targets 
```

```python
import pandas as pd
dataframe = pd.DataFrame(Features, columns=breast_cancer_wisconsin_diagnostic.features)
dataframe['target'] = Target
dataframe['ID'] = Id

dataframe
```

```python
# Convert Target into Binary
dataframe['target'].replace({'B': 0, 'M': 1}, inplace=True)
dataframe
```

```python
# Looking for value counts
dataframe['target'].value_counts()
```

```python
# Fixing class imblance using SMOTE
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='minority')
```

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import tensorflow as tf
import imblearn
import keras_tuner as kt
```

```python
# Separate data into Features and Target
y = dataframe.target.values
X = dataframe.drop(columns=['ID','target'])

# Splitting into Train and Test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
# Fix class imbalances in training data
X_smote, y_smote = smote.fit_resample(X_train, y_train)
```

```python
scaler = StandardScaler()

# X_scaler = scaler.fit(X_smote)
X_scaler = scaler.fit(X_train)

# X_train_scaled = X_scaler.transform(X_smote)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
def create_model(hp):
    nn_model = tf.keras.models.Sequential()

    # Allow kerastuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation',['relu','sigmoid'])
    
    # Allow kerastuner to decide number of neurons in first layer
    nn_model.add(tf.keras.layers.Dense(units=hp.Int('first_units',
        min_value=1,
        max_value=30,
        step=5), activation=activation, input_dim=30))

    # Allow kerastuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 1, 5)):
        nn_model.add(tf.keras.layers.Dense(units=hp.Int('units_' + str(i),
            min_value=1,
            max_value=30,
            step=5),
            activation=activation))
    
    nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    # Compile the model
    nn_model.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])
    
    return nn_model
```

```python
tuner = kt.Hyperband(
    create_model,
    objective="val_accuracy",
    max_epochs=50,
    hyperband_iterations=5)
```

```python
tuner.search(X_train_scaled,y_train,epochs=20,validation_data=(X_test_scaled,y_test))
```

```python
top_hyper = tuner.get_best_hyperparameters()
for param in top_hyper:
    print(param.values)
```

```python
top_model = tuner.get_best_models()
for model in top_model:
    model_loss, model_accuracy = model.evaluate(X_test_scaled,y_test,verbose=2)
    print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Get the top hyperparameters
top_hyper = tuner.get_best_hyperparameters()

# Build and train a model using the top hyperparameters
best_model = tuner.hypermodel.build(top_hyper[0])
best_model.fit(X_train_scaled, y_train, epochs=20, validation_data=(X_test_scaled, y_test))
```

```python
X_scaled = X_scaler.transform(X)

predictions = best_model.predict(X_scaled)
```

```python
predicted_labels = (predictions > 0.5).astype(int)
predicted_df = pd.DataFrame(predicted_labels)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_curve, auc

conf_matrix = confusion_matrix(y, predicted_labels)

class_report = classification_report(y, predicted_labels)

accuracy = accuracy_score(y, predicted_labels)

# print("Confusion Matrix:\n", conf_matrix)
print("\nClassification Report:\n", class_report)
print("\nAccuracy Score:", accuracy)
```
